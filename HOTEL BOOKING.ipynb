```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("data//hotel_booking.csv")
```

```python
df
```

```python
df.info()
```

**WHAT IS THE  DISTRIBUTION OF "CUSTOMER TYPE" ACCORDING TO CITY HOTEL AND RESORT HOTEL?**

```python
plt.figure(figsize=(10,10))
sns.histplot(data = df, x = "hotel",palette = "Set1", hue = "customer_type", multiple = "stack")
plt.title("Types of Booked Hotels")
plt.xlabel("Hotel Type")
plt.ylabel("Number of Bookings")
```

**NO.OF STAYS IN WEEKEND NIGHTS vs NO.OF STAYS IN WEEK NIGHTS**

```python
df['stays']= df['stays_in_weekend_nights']
plt.figure(figsize=(18,5))
sns.countplot(x=df['stays'], color='r')
plt.title("Stays in Weekend Nights")
plt.show()

df['stays']=df['stays_in_week_nights']
plt.figure(figsize=(18,5))
sns.countplot(x=df['stays'], color='b')
plt.title("Stays in Week Nights")
plt.show()
```

**WHAT IS THE PECENTAGE OF CANCELLATION?**

```python



plt.figure(figsize=(10,10))
df['is_canceled'].value_counts().plot.pie(labels=['Not Canceled', 'Canceled'],autopct="%1.1f%%",explode=(0.1,0))
plt.title("Cancellation and non Cancellation")
plt.show()
```

**WHICH TYPE OF HOTEL IS MOSTLY PREFERED BY THE GUESTS?**

```python
df['hotel'].value_counts().plot.pie(explode=[0.05, 0.05], autopct='%1.1f%%', shadow=True, figsize=(10,8),fontsize=20)   
plt.title('Pie Chart for Most Preffered  Hotel')
```

**WHAT IS THE PERCENTAGE OF REPEATED GUESTS?**

```python
df['is_repeated_guest'].value_counts().plot.pie(labels=['New Guest', 'Repeated Guest'],explode=(0.05,0.05),autopct='%1.1f%%',shadow=True,figsize=(12,8),fontsize=20)

plt.title(" Percentgae (%) of repeated guests")
```

**WHAT IS THE PERCENTAGE DISTRIBUTION OF "CUSTOMER TYPE"?**

```python
df['customer_type'].value_counts().plot.pie(shadow=True,autopct='%1.1f%%',figsize=(12,8),fontsize=15,)
labels=df['customer_type'].value_counts()
plt.title('% Distribution of Customer Type')
plt.legend(bbox_to_anchor=(0.85, 1), loc='upper left', labels=labels)
```

**WHAT IS THE PERCENTAGE DISTRIBUTION OF REQUIRED_CAR_PARKING_SPACES?**

```python
df['required_car_parking_spaces'].value_counts().plot.pie(explode=[0.05]*5,autopct='%1.1f%%',shadow=False,figsize=(12,8),fontsize=15,labels=None)
labels=df['required_car_parking_spaces'].value_counts().index
plt.title('% Distribution of required car parking spaces')
plt.legend(bbox_to_anchor=(0.85, 1), loc='upper left', labels=labels)
```

 **WHICH TYPE OF FOOD IS MOSTLY PREFERRED BY THE GUESTS?**

```python
plt.figure(figsize=(18,8))
sns.countplot(x=df['meal'])
plt.xlabel('Meal Type')
plt.ylabel('Count')
plt.title("Preferred Meal Type")
```

**WHAT IS PERCENTAGE DISTRIBUTION OF DEPOSITE TYPE ?**



```python
df['deposit_type'].value_counts().plot.pie(explode=(0.05,0.05,0.03),autopct='%1.1f%%',shadow=False,figsize=(14,8),fontsize=20,labels=None)
plt.title("% Distribution of deposit type")
labels=df['deposit_type'].value_counts().index.tolist()
plt.legend(bbox_to_anchor=(0.85, 1), loc='upper left', labels=labels)
```

**FROM WHICH COUNTRY THE MOST GUESTS ARE COMING?**



```python
country_df=df['country'].value_counts().reset_index().rename(columns={'index': 'country','country': 'count of guests'})[:10]
```

```python
country_df
```

```python
plt.figure(figsize=(20,8))
sns.barplot(x=country_df['country'],y=country_df['count of guests'])
plt.xlabel('Country')
plt.ylabel('Number of guests',fontsize=12)
plt.title("Number of guests from diffrent Countries")
```

**WHICH IS THE MOST PREFERRED ROOM TYPE BY THE CUSTOMERS?**



```python
plt.figure(figsize=(18,8))
sns.countplot(x=df['assigned_room_type'],order=df['assigned_room_type'].value_counts().index)
plt.xlabel('Room Type')
plt.ylabel('Count of Room Type')
plt.title("Most preferred Room type")
```

**WHICH YEAR HAD THE HIGHEST BOOKINGS?**

```python
plt.figure(figsize=(12,8))
sns.countplot(x=df['arrival_date_year'],hue=df['hotel'])
plt.title("Year Wise bookings")
```

**WHICH HOTEL TYPE HAS THE HIGHEST ADR?**



```python
grup_by_hotel=df.groupby('hotel')
highest_adr=grup_by_hotel['adr'].mean().reset_index()
plt.figure(figsize=(10,8))
plt.title("Avg ADR of each Hotel type")
sns.barplot(x=highest_adr['hotel'],y=highest_adr['adr'])
```
